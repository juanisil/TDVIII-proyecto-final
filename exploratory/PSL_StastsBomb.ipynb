### Sample Comp: La Liga 2015/2016

```python
ll_events = sb.competition_events(
    country="Spain",
    division="La Liga",
    season="2015/2016",
    # include_360_metrics=True,
)
```

```python
ll_events
```

```python
barca_events = ll_events[ll_events.team == "Barcelona"]
barca_events
```

```python
list(barca_events.columns)
```

```python
barca_events.type.value_counts()
```

```python
tactics = list(barca_events[barca_events.tactics.notnull()].tactics.values)
tactics[0]["lineup"]
```
